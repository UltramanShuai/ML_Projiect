## Data Collection

## Data description

The row data contains 8 attributes.

- job: The job title. (String)

- Company: The name of the company. (String)

- Salary: Salary information. (String)

- Work_Position: The location (suburb) of the job. (String)

- Post_time: The time of the job being posted . (String) <br>
This column could not help our analysis since the post time changes automatically daily to the present day.

- Require: The requirements of the job. (String)

- Describe: The additional information that may or may not help our analysis from the job posting. (String)

- Type: The "tags" associated with the job or the position, eg. industry, sector etc. (String)

## Data pre-processing

The raw data was very complicated and chaotic to process. To follow the "garbage in, garbage out" rule, data collection and data pre-processing are the very important steps in Machine Learning process and which will take most of our time. 

For data pre-processing, `pandas`, `numpy` and `re` package were used.

```python
import pandas as pd
import numpy as np
import re
```

#### Read data and drop item without salary

The data was saved as a `.csv` file. However, the language and the format of the data can not be read directly by using `pd.read_csv()`, which means we should use an alternative  way. The following code shows the process of data importation and re-formating. 

```python
f=open("job1.csv")
df=pd.read_csv(f)
df["job"]=df.index
df.columns=[ 'Company', 'Salary', 'Work_Position', 'Post_time',
       'Require', 'Describe',"type",'job']
df.head()
```

The data are really messy especially in `Require`, `Describe` and `Type`. All of those attributes contain `[]` which represent `nan` value, should be change to `nan`.

```python
# relapce nan
df=df.replace('[]',np.nan)
```

Since the raw data contains duplicates, we should remove the duplicates in order to perform the modeling process.

```python
# drop duplicate item
df.drop_duplicates(inplace=True)
```

The ideal results of the project has direct association with the salary of the job, which means all data without salary should be dropped.

```python
# drop items witout salary
df.drop(df[df['Salary'].isnull()].index.tolist(),inplace=True)
# re range the index
df.index=range(len(df))
df.head(3)
```

After processing, it is more adapt for further analysis.

```python
df.info()
```

There are still 74660 items here. 

```python
df["job"].value_counts().describe()
```

There are 51788 different jobs founded, which means more than half of the job postings are under different jobs.

#### Dealing with salary

Based on the raw data, the attibutes in `salary` were not presented as a numeric value  which they cannot be recognized by python.

Therefore, what we should do now is to unify the units of the attibutes and transfer the attributes into numeric values.

```python
salary_first=df["Salary"].str.split('/',expand=True)[0].str.split('-',expand=True)[0]
salary_last=df["Salary"].str.split('/',expand=True)[0].str.split('-',expand=True)[1]
```

`is_number` function transforms numeric data as numeric, not as `nan` which can be deleted easily latter.

```python
def is_number(num):
    try: 
        float(num)
        return float(num)
    except ValueError: 
        return np.nan
```

Extract useful information to compute monthly salary

```python
df["salary_first"]=[is_number(x) for x in salary_first]
df["salary_second"]=[is_number(re.findall(u"[^\u4e00-\u9fa5]+",x)[0]) if x else x for x in salary_last]
df["salary_unit"]=[re.findall(u"[\u4e00-\u9fa5]+",x)[0] if x else x for x in salary_last]
df["salary_time"]=df["Salary"].str.split('/',expand=True)[1]
```

The price units founded were 10 thousand and thousand.

```python
df["salary_unit"].value_counts()
```

Replace the price unit by numeric value.

```python
df["salary_unit"]=df["salary_unit"].replace("万",10000)
df["salary_unit"]=df["salary_unit"].replace("千",1000)
```

The time units contain month, year, day and hour.

```python
df["salary_time"].value_counts()
```

replace time unit into numeric based on monthly.

```python
df["salary_time"]=df["salary_time"].replace("月",1)
df["salary_time"]=df["salary_time"].replace("年",0.083)
df["salary_time"]=df["salary_time"].replace("天",30)
df["salary_time"]=df["salary_time"].replace("小时",240)
```

Drop the items without useful information and re-range the index of data. Still, 73938 items retrain.

```python
df.drop(df[df['salary_first'].isnull()].index.tolist(),inplace=True)
df.index=range(len(df))
df.drop(df[df['salary_time'].isnull()].index.tolist(),inplace=True)
df.index=range(len(df))
df.info()
```

Compute the numeric monthly salary.

```python
df["salary_low"]=df["salary_first"]*df["salary_unit"]*df["salary_time"]
df["salary_high"]=df["salary_second"]*df["salary_unit"]*df["salary_time"]
df["salary_mean"]=(df["salary_low"]+df["salary_high"])/2
```

Drop the original salary column.

```python
df=df.drop(["Salary","salary_first","salary_second","salary_unit","salary_time"],axis=1)
df.head(2)
```

looks great!

### Machine learning

In this part, we will try to predict if the salary will higher than 30000 per month based on the description, type and requirement.

The results from this part can help job seekers to: know the requirements of these highly paid jobs and set goals for themselves; or based on the requirements and job functions to set their career path.

`get_list()` function can return several top frequent words based on the list  which was sent and the number of words the return list should contain. It also filters the meaningless words for our task.

```python
# send a list with dupicate itme, return the list contains most occuray itme. the number of return items can be define.
from collections import Counter
def get_list(desire_list,num):
    c=Counter(desire_list)
    
    def delete_word(word):
        B_R=str(word)!="nan" and str(word)!="" and not str(word).startswith("北京") and not str(word).startswith("01-") and not str(word).startswith("招") and not str(word).startswith("五险")
        return B_R
        
    return [c.most_common(num)[i][0] for i in range(num) if delete_word(c.most_common(num)[i][0])]
```

`add_column()` function can help us find if the particular attribute of data contains the keywords and send to a model dateframe.

```python
def add_column(list_to_columns,from_columns):
    for item in list_to_columns:
        df_model[item]=df[from_columns].str.contains(item)
        df_model[item].replace(np.nan,False,inplace=True)
```

First of all, we build `df_model` set which receives  the data from `add_clumn` function.

The training data in this set contains every observation if they contain top 100 frequency keywords from requiring, description and requirement.

```python
df_model = pd.DataFrame()
add_column(get_list(type_list,100),"type")
add_column(get_list(describe_list,100),"Describe")
add_column(get_list(require_list,100),"Require")
```

Except for using over-sampling or under-sampling methods to deal with data, we can also tune the parameters to make the algorithm suit for our task.

#### Tune parameters

By using `GridSearchCV`, we can simply run several models with different parameters and find the one to be the best result.

One thing needs to mention is we need to use balanced_accuracy as our score to measure our model.(imbalance data)

```python
from sklearn.model_selection import GridSearchCV 
param_grid = [ {'n_estimators': [10, 20, 100],
                'criterion': ["gini", "entropy"],
                'max_features': ["auto", "log2"]}, 
              {'bootstrap': [False,True], 'min_samples_leaf': [1,2,4,10],'class_weight':["balanced_subsample","balanced"]}] 
forest_clf = RandomForestClassifier()
grid_search = GridSearchCV(forest_clf, param_grid, cv=3, scoring='balanced_accuracy')
grid_search.fit(train_X, train_y)
```

There are 28 model was tested and comes out with different scores.

```python
cvres = grid_search.cv_results_ 
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]): 
    print(mean_score, params)
```

The best model with particular parameters.

```python
grid_search.best_estimator_
```

The best group of parameter we set.

```python
grid_search.best_params_
```

Make the model comes out the best parameters as our final model.

```python
final_model=grid_search.best_estimator_
```

```python
cross_val_score(final_model, train_X, train_y, cv=3, scoring="accuracy")
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label) 
    plt.plot([0, 1], [0, 1], 'k--') 
    plt.axis([0, 1, 0, 1]) 
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
```

```python
plot_roc_curve(fpr_forest, tpr_forest, "Random Forest")
plt.legend(loc="lower right")
plt.show()
```

#### ROC plot shows the model performance. The result is not that bad consider the raw data was really messy.
